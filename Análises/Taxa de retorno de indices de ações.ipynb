```python
import numpy as np
import pandas as pd
from pandas_datareader import data as wb
import matplotlib.pyplot as plt
import yfinance as yf
yf.pdr_override()
```

```python
# os índices de ações que serão usados: S&P500(GSPC), NASDAQ(IXIC), Índice Alemão DAX(GDAXI), NASDAQ(FTSE) 
# O ^antes de cada símbolo e para indicar que estamos usando indices de ações 
```

```python
tickers = ['^GSPC', '^IXIC', '^GDAXI', '^FTSE']
ind_data = pd.DataFrame()

for t in tickers:
    ind_data[t] = wb.DataReader(t, data_source = 'yahoo', start='1997-1-1')['Adj Close']
```

```python
ind_data.head()
```

```python
#Os valores dos indices são mais altps em relação aos valores dos índices ajustados das empresas que vimos anteriormente
#Vmos plotar os dados em um gráfico normaliando para a base 100
```

```python
(ind_data / ind_data.iloc[0] * 100).plot(figsize=(15,6))
```

```python
#Perto do ano 2000 temos um aumento geral no valor dos indices, isso porque entre 1995 e 2001 houve a bolha da internet e os mercados
# de ações de países desenvolvidos como EUA foram impulsionados pelo desenvolvimento de um valor significativo do valor das empresas
# de internet
```

```python
#Calculando o retorno simples dos números
```

```python
ind_returns = (ind_data / ind_data.shift(1)) - 1
```

```python
ind_returns.tail()
```

```python
#O resultado desssa célula mostra os retornos anuais
```

```python
annual_ind_returns = ind_returns.mean() * 250
annual_ind_returns
```

```python
#Todos os retornos são positivos> A implicação é que a média das empresas listas sobre os quatro indices garantiu aos investidores
#uma taxa positiva de retorno nas ultimas decadas. Por esse periodo a nasdac superou facilmente o SP500
```
